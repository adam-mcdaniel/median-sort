```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Read in the quicksort.csv file
df = pd.read_csv('quicksort.csv')

plt.figure(figsize=(12, 6))

# Plot a line for each `r` value, for all the cutoffs
for r in df['r'].unique():
    sns.lineplot(x='cutoff', y='median_time (seconds)', data=df[df['r'] == r], label=f'r={r}')

# Set the title and labels
plt.title('Quicksort with different cutoffs')
plt.xlabel('Cutoff')
plt.ylabel('Median time (seconds)')
plt.show()
```

```python
# Show a bar graph of the median runtimes for each r value
plt.figure()
sns.barplot(x='r', y='median_time (seconds)', data=df)
plt.title('Median runtime for each r value')
plt.xlabel('r')
plt.ylabel('Median time (seconds)')
plt.ylim(0.3, 0.4)
plt.yticks([0.3, 0.325, 0.35, 0.375, 0.4, 0.425])
plt.show()
```
